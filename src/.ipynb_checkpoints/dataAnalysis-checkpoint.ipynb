```python
import matplotlib.pyplot as plt
import os
import glob
import pandas as pd
```

**Data Analysis**

This data is represented in 6 columns.  These are time, power, oxygen, cadence, HR, and RF.

**Time**

Time is represented in seconds of activity.  There were many trials that make up the overall data, and most trials ended around 2,400 seconds although some may differ.

```python
data = pd.read_csv("../data/concatedData/combined_csv.csv")
data.size
```

Here we can see we have 1,165,986 datapoints to work with corresponding to 1 second each.

**Power**

The power column is represented as Watts.

```python
column = data['Power']
print(column.max())
print(column.min())
```

Here we can see the maximum power outputted within all the data is 899 watts, and the minimum is 0(Not moving).

```python
plt.scatter(data['time'], data['Power'])
plt.title("Power vs Time")
plt.xlabel("Seconds")
plt.ylabel("Watts")
plt.show()
```

Above is all the data when plotted power vs time.  Better trends can be seen when showing individual trials.

```python
dataTrial1 = pd.read_csv("../data/rawData/sbj_1_I.csv")
plt.plot(dataTrial1['time'], dataTrial1['Power'])
plt.title("Power vs Time")
plt.xlabel("Seconds")
plt.ylabel("Watts")
plt.show()
```

Here we can also see the type of workout that was done by this participant.  Below is another example of a workout in the dataset.

```python
dataTrial3 = pd.read_csv("../data/rawData/sbj_1_II.csv")
plt.plot(dataTrial3['time'], dataTrial3['Power'])
plt.title("Power vs Time")
plt.xlabel("Seconds")
plt.ylabel("Watts")
plt.show()
```
